```python
import json
import os, glob
import pandas as pd
import pandas.io.json as pdjson
import seaborn as sns
```

```python
data_frames = []

for file in glob.glob("*.json"):
    with open(file) as json_file:
        data = json.load (json_file)
        df = pdjson.json_normalize(data).head(50)
        df['variant'] = file.replace(".json","")
        data_frames.append(df)
```

```python
df = pd.concat (data_frames)
df.count()
```

```python
g = sns.catplot (x='name', y='time_secs', hue='variant', data = df, kind ='bar', aspect=4)
g.set_xticklabels(rotation=90)
```

```python
g = sns.catplot (x='name', y='sys_time_secs', hue='variant', data = df, kind ='bar', aspect=4)
g.set_xticklabels(rotation=90)
```
